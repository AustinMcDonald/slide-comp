```python
end = 300
DataFiles = glob.glob(Dir+'*')
DataFiles.sort()
DataFiles = DataFiles[0:end]
```

```python
Row    = int(250)
Col    = Row
Yindex = int(256)
Xindex = int(256)

eOffset, eCoeff = PE_Vals(DataFiles[0])
Test = Image_Converter(DataFiles[0], eOffset, eCoeff, Xindex, Yindex, Row, Col)
Shape = Test.shape[0]

plt.figure(figsize=(7,7))

r = 130
Xoff = 250
Yoff = 320

X_circle, Y_circle = Unique_Circle(r, Xoff, Yoff)
keep = np.where(Y_circle<Shape)
Y_circle = Y_circle[keep]
X_circle = X_circle[keep]
keep = np.where(X_circle<Shape)
Y_circle = Y_circle[keep]
X_circle = X_circle[keep]

plt.scatter(X_circle,Y_circle,color='r',lw=2,alpha=0.4,s=1)

plt.imshow(Test)
plt.xlim(0,500)
plt.ylim(0,500)
plt.show()
```

```python
FreqCut=0.03
FreqCutWidth=0.04
FilterArray = FrequencyFilterFunction(Shape,FreqCut,FreqCutWidth)
```

```python
enddd  = np.zeros(Shape**2).reshape((Shape,Shape))
for img in DataFiles[end-10:end]:
    eOffset, eCoeff = PE_Vals(img)
    ReducedImage = Image_Converter(img, eOffset, eCoeff, Xindex, Yindex, Row, Col)
    enddd += ReducedImage    
enddd = enddd/10

TEST  = np.zeros(Shape**2).reshape((Shape,Shape))
TEST2 = np.zeros(Shape**2).reshape((Shape,Shape))

r_cut  = np.sqrt((X_circle-Xoff)**2+(Y_circle-Yoff)**2).max()
r_vals = Test[X_circle,Y_circle]

for r in range(Test.shape[0]):
    for c in range(Test.shape[0]):
        rr = np.sqrt((r-Yoff)**2+(c-Xoff)**2)
        if rr<=r_cut:
            TEST[r,c]=1
        else:
            TEST2[r,c]=np.mean(r_vals)
            
ypos, xpos  = np.indices(enddd.shape) 
zpos = enddd*TEST+TEST2

fit_params, fit_errors, chi_squared = Fit_2D_Gaussian(zpos)
print("chi squared / dof = ",chi_squared)

x = np.arange(Shape)
y = np.arange(Shape)
xy_mesh = np.meshgrid(x, y)
FIT = gaussian_2d(xy_mesh,*fit_params)
FIT = FIT.reshape(Test.shape)
#endddF = zpos-FIT.reshape(Test.shape)
#endddF[endddF<0]=0
```

```python
datamin  = 0
datamax  = 320
numbins  = 320
bins     = np.linspace(datamin, datamax, numbins)
HIST     = np.zeros(numbins-1, dtype='int32')

for img in DataFiles:
    HIST_VAL = []
    eOffset, eCoeff = PE_Vals(img)
    ReducedImage = Image_Converter(img, eOffset, eCoeff, Xindex, Yindex, Row, Col)
    ReducedImage = ReducedImage-FIT
    ReducedImage[ReducedImage<0]=0
    ReducedImage = FFT_Filter(ReducedImage, FilterArray)
    r_cut  = np.sqrt((X_circle-Xoff)**2+(Y_circle-Yoff)**2).max()
    r_vals = ReducedImage[X_circle,Y_circle]

    for r in range(ReducedImage.shape[0]):
        for c in range(ReducedImage.shape[0]):
            rr = np.sqrt((r-Yoff)**2+(c-Xoff)**2)
            if rr<=r_cut:
                HIST_VAL.append(ReducedImage[r,c])
    htemp, jnk = np.histogram(HIST_VAL, bins)
    HIST += htemp

np.save(name,HIST)
```
