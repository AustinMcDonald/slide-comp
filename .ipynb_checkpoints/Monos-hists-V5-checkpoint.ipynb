```python
from scipy import interpolate
```

```python
con = 1.1
xnew = np.arange(0, 200, 0.1)
plt.figure(figsize=(8,8))

one = Baspik06/np.sum(Baspik06)
#one[one<1e-8] = 1e-8
two = Baspik05/np.sum(Baspik05)
#two[two<1e-8] = 1e-8
f1 = interpolate.interp1d(bincenters, one)
f2 = interpolate.interp1d(bincenters*con, two)
this = (f2(xnew) - f1(xnew))/f1(xnew)
plt.plot(xnew,this ,color='k',label="Barium spiked 06-05 ",lw=3)

one = Bafree06/np.sum(Bafree06)
#one[one<1e-8] = 1e-8
two = Bafree05/np.sum(Bafree05)
#two[two<1e-8] = 1e-8
f1 = interpolate.interp1d(bincenters, one)
f2 = interpolate.interp1d(bincenters*con, two)
this = (f2(xnew) - f1(xnew))/f1(xnew)
plt.plot(xnew,this ,color='r',label="Barium free 06-05 ",lw=3)

one = Poly06/np.sum(Poly06)
#one[one<1e-8] = 1e-8
two = Poly05/np.sum(Poly05)
#two[two<1e-8] = 1e-8
f1 = interpolate.interp1d(bincenters, one)
f2 = interpolate.interp1d(bincenters*con, two)
this = (f2(xnew) - f1(xnew))/f1(xnew)
plt.plot(xnew,this ,color='b',label="Poly 06-05 ",lw=3)

one = Glass06/np.sum(Glass06)
#one[one<1e-8] = 1e-8
two = Glass05/np.sum(Glass05)
#two[two<1e-8] = 1e-8
f1 = interpolate.interp1d(bincenters, one)
f2 = interpolate.interp1d(bincenters*con, two)
this = (f2(xnew) - f1(xnew))/f1(xnew)
plt.plot(xnew,this ,color='g',label="glass06-05 ",lw=3)


plt.legend(loc='upper left',fontsize=16)
plt.xlabel('Pixel intensity',fontsize = 18)
plt.ylabel('Counts',fontsize = 18)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlim(0,140)
plt.ylim(-2,12)
#plt.semilogy()
plt.grid(True)
plt.tight_layout()

#plt.savefig('bar10.png',dpi=400)
plt.show()
```
