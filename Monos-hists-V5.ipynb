```python
import numpy as np
import glob

# for plotting
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
#from matplotlib.colors import LogNorm
%matplotlib inline


from scipy import interpolate
```

```python
ybaE  = np.load("hist-Exp-wba.npy")
nbaE  = np.load("hist-Exp-woba.npy")
glaE  = np.load("hist-Exp-glass.npy")
polE  = np.load("hist-Exp-poly.npy")
```

```python
ybaD  = np.load("hist-Des-wba.npy")
nbaD  = np.load("hist-Des-woba.npy")
glaD  = np.load("hist-Des-glass.npy")
polD  = np.load("hist-Des-poly.npy")
```

```python
bincenters = np.arange(len(ybaE))
```

```python
plt.figure(figsize=(8,8))

plt.plot(bincenters,ybaE ,color='k',label="Barium spiked",lw=3)
plt.plot(bincenters,nbaE ,color='r',label="Barium free",lw=3)
plt.plot(bincenters,polE ,color='b',label="poly",lw=3)
plt.plot(bincenters,glaE ,color='g',label="glass",lw=3)


plt.plot(bincenters,ybaD ,color='k',lw=3,linestyle='--')
plt.plot(bincenters,nbaD ,color='r',lw=3,linestyle='--')
plt.plot(bincenters,polD ,color='b',lw=3,linestyle='--')
plt.plot(bincenters,glaD ,color='g',lw=3,linestyle='--')

plt.legend(loc='upper right',fontsize=16)
plt.xlabel('Pixel intensity',fontsize = 18)
plt.ylabel('Counts',fontsize = 18)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlim(0,160)
plt.semilogy()
plt.grid(True)
plt.tight_layout()

#plt.savefig('bar10.png',dpi=400)
plt.show()
```
